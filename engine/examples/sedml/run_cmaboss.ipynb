```python
import maboss
```

```python
m = maboss.loadSBML("cell_fate.sbml", cmaboss=True)
m.network.set_output(["Survival", "NonACD", "Apoptosis"])
m.update_parameters(max_time=100, time_tick=0.1, sample_count=10000)
```

```python
r = m.run()
r.plot_piechart()
```

```python
m2 = m.copy()
```

```python
m2.network.set_istate("TNF", [0, 1])
```

```python
r2 = m2.run()
r2.plot_piechart()
```

```python
r2.plot_trajectory()
```

```python
m3 = m2.copy()
m3.mutate("IKK", "ON")
m3.mutate("cFLIP", "ON")
```

```python
r3 = m3.run()
r3.plot_piechart()
```
